# 5. Post-devCellPy Analysis in R

The output of the PredictOne and PredictAll modes of devCellPy includes a CSV file containing the cell barcodes and the corresponding celltype labels. Here we show how to import the cell predictions made by the devCellPy into an object in Seurat. 

This CSV file can be added to the metadata of the test dataset Seurat Object for downstream visualization. Since the CSV file contains cell barcodes, the annotations can be easily added to each layer of annotation. 

## Example of Prediction Output

# Load the CSV output for the prediction of a query dataset

galdos_devcellpy_pred = read.csv("/path/to/devcellpy_predictAll_20220615075059/predictionall_reject0.5.csv", header = FALSE)

colnames(lietal_devcellpy_pred) = c('Barcodes', "Celltype1', 'Celltype2', 'Celltype3')
rownames(lietal_devcellpy_pred) = lietal_devcellpy_pred$Barcodes

# Add Celltype Predictions to Seurat Object

lietal_seurat_object$devcellpy_Celltype1 = lietal_devcellpy_pred$Celltype1
lietal_seurat_object$devcellpy_Celltype2 = lietal_devcellpy_pred$Celltype2
lietal_seurat_object$devcellpy_Celltype3 = lietal_devcellpy_pred$Celltype3

DimPlot(lietal_seurat_object, group.by = 'devcellpy_Celltype1')
DimPlot(lietal_seurat_object, group.by = 'devcellpy_Celltype2')
DimPlot(lietal_seurat_object, group.by = 'devcellpy_Celltype3')
